```python
X = df[['number of bedrooms', 'number of bathrooms','living area','condition of the house','Number of schools nearby' ]]
Y = df[['Price']]
```

```python
X
```

```python
Y
```

```python
%pip install scikit-learn
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,Y,  test_size= 0.2)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# 1. Initialize the model
model = LinearRegression()

# 2. Train the model (Fitting)
model.fit(X_train, y_train)

# 3. Make predictions on the test set
predictions = model.predict(X_test)

# 4. Check the performance
print(f"R^2 Score: {r2_score(y_test, predictions)}")
print(f"MSE: {mean_squared_error(y_test, predictions)}")
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import GridSearchCV

# 1. Define the parameters to test
param_grid = {
    "criterion": ["squared_error", "friedman_mse", "absolute_error"],
    "splitter" : ["best", "random"],
    "max_depth" : [None, 10, 20, 30, 40, 50],
    "min_samples_split" : [2, 5, 10],
    "min_samples_leaf" : [1, 2, 4]
}

# 2. Initialize the model
tree_model = DecisionTreeRegressor()

# 3. Setup the Grid Search
# This will try every combination of parameters defined above
grid_tree = GridSearchCV(estimator=tree_model, param_grid=param_grid)

# 4. Train the models on your training data
grid_tree.fit(X_train, y_train)
```

```python
 tree_preds = grid_tree.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error
from sklearn.linear_model import LinearRegression

# 1. Predict using the GridSearch Decision Tree
tree_preds = grid_tree.predict(X_test)

# 2. Evaluate Decision Tree performance
mse_tree = mean_absolute_error(y_test, tree_preds)
print(mse_tree)

# 3. Initialize and train a Linear Regression model
lr = LinearRegression()
lr.fit(X_train, y_train)

# 4. Make predictions with Linear Regression
predslr = lr.predict(X_test)
```

```python
mean_absolute_error(y_test,predslr)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV

# Initialize the model
rfrmodel = RandomForestRegressor()

# Define the parameter grid
param_gridrfr = {
    "max_depth": [5, 10, 15],
    "n_estimators": [2, 3, 4, 5, 6, 7, 8, 9, 10]
}

# Setup GridSearchCV
gridrfr = GridSearchCV(rfrmodel, param_gridrfr)

# Fit the model to find the best parameters
gridrfr.fit(X_train, y_train.values.ravel())
```

```python
gridrfr.best_params_
```

```python
import joblib
from sklearn.metrics import mean_absolute_error

# 1. Generate final predictions using the tuned Random Forest
rfrpredictions = gridrfr.predict(X_test)

# 2. Calculate the Mean Absolute Error
# This tells you, on average, how many dollars off your price predictions are
error = mean_absolute_error(y_test, rfrpredictions)
print(error)

# 3. View the model object structure
gridrfr

# 4. Save (serialize) your model to a file named 'model.pkl'
joblib.dump(gridrfr, "model.pkl")
```
